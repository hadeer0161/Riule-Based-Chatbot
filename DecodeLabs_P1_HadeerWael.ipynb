```python
# Hadeer Wael AbdelAziz
# Project 1: Rule-Based AI Chatbot

def sanitize_input(raw_text):
    return raw_text.lower().strip()

def process_and_respond(clean_text):
    # Greeting Intents
    if clean_text in ["hello", "hi", "hey", "greetings"]:
        return "Bot: Hello! Welcome to the team at DecodeLabs. How can I help you today?"

    # Project 1 info Intents
    elif "project 1" in clean_text or "rule-based" in clean_text:
        return "Bot: Project 1 is a Rule-Based AI Chatbot focused on control flow, logic, and deterministic guardrails."

    # DecodeLabs info Intents
    elif "decodelabs" in clean_text:
        return "Bot: DecodeLabs helps you build a real-world portfolio to showcase your foundational AI concepts!"

    # Help Intent
    elif "help" in clean_text:
        return "Bot: I can answer questions about Project 1, DecodeLabs, or explain rule-based logic. Try asking me about them!"

    # Fallback response for unmatched input (Deterministic filter)
    else:
        return "Bot: I'm not sure I understand that yet. Try expanding my vocabulary with more nested conditions!"

def start_chatbot():
    print("=" * 60)
    print("         DECODELABS RULE-BASED AI CHATBOT INITIALIZED         ")
    print("               Type 'exit', 'quit', or 'bye' to stop.         ")
    print("=" * 60)

    while True:
        # Phase 1: Capture Raw Input (Targeting the exact syntax from your kit)
        raw_input = input('You: ')

        # Sanitization & Normalization Process
        clean_input = sanitize_input(raw_input)

        # Handle Exit Commands
        if clean_input in ["exit", "quit", "bye"]:
            print("Bot: Goodbye! Good luck building your AI milestones at DecodeLabs!")
            break

        # Phase 2 & 3: Process input and generate the feedback output loop
        response = process_and_respond(clean_input)
        print(response)
        print("-" * 40)

# Start the continuous simulation loop
start_chatbot()
```
